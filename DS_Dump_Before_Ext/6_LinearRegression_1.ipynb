```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(42)

#rand generates random float values between 0 and 1 -- [0,1) Uniform Distribution
#multiplying it with 5 gives 100 random numbers between 0 and 5
#with 100 rows and 1 column
x = 5 * np.random.rand(100,1) 
#Creates y values with linear relationship like y=mx+c
#for each x values multiplies it with 3 and adds 4
#then adds a random noise
y = 4 + 2 * x + np.random.randn(100, 1)
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
plt.scatter(x,y)
plt.title('Basic Scatter Plot')
plt.show()
```

```python
x2 = np.array([[0],[5]]) #
y2 = reg.predict(x2)
```

```python
plt.scatter(x,y,c='g',alpha=0.5)
plt.plot(x2,y2,c='r')
plt.title('Scatter Plot with Line')
plt.show()
```
