```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(41)

x = np.random.randint(10,30,size=100)
y = x + np.random.randint(5,10,size=100)
x = x.reshape(-1, 1)

reg = LinearRegression()
reg.fit(x,y)

x2 = np.array([10, 30]).reshape(-1, 1)
y2 = reg.predict(x2)

plt.scatter(x,y)
plt.plot(x2,y2,c='r')
plt.show()
```
