```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
data = load_iris()
print(data['feature_names'])
print(data['target_names'])
```

```python
x = data.data
y = data.target
```

```python
reg = KMeans(n_clusters=3,random_state=1, n_init='auto') #Training a regression model
reg.fit(x) #training with features
```

```python
#Assigning Cluster centers and labels into variables
Centers = reg.cluster_centers_
Labels = reg.labels_
# print(Centers)
# print(Labels)
# print(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=Labels,cmap='viridis') 
plt.scatter(Centers[:,0],Centers[:,1],color='red',marker='o',s=1000,alpha=0.3)
plt.show()
```

```python
pred = reg.predict([[4.8,3.4,1.6,0.2]])
print(data.target_names[pred])
```
