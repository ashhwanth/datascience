```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.naive_bayes import GaussianNB
```

```python
np.random.seed(42)
data = {'Habitat':np.random.choice(['Land','Sea','Water'],size=25),
        'Legs':np.random.choice([2,4,6],size=25),
        'Wings':np.random.choice(['Yes','No'],size=25),
        'Fur':np.random.choice(['Yes','No']),
        'Type':np.random.choice(['Animal','Bird','Fish'],size=25)}

df = pd.DataFrame(data)
```

```python
df['Habitat'] = df['Habitat'].map({'Land':0,'Sea':1,'Water':2})
df['Wings'] = df['Wings'].map({'Yes':0,'No':1})
df['Fur'] = df['Fur'].map({'Yes':0,'No':1})

x = df[['Habitat','Legs','Wings','Fur']]
y = df['Type']
```

```python
#DecisionTree
clf = DecisionTreeClassifier()
clf.fit(x,y)
```

```python
pred = clf.predict([[0,4,1,0]])
print(pred)
```

```python
plt.figure.figsize=(18,6)
plot_tree(clf,feature_names=x.columns,class_names=['Animal','Bird','Fish'],filled=True,rounded=True)
plt.show()
```

```python
#NaiveBayes

clf2 = GaussianNB()
clf2.fit(x,y)
```

```python
pred2 = clf2.predict([[0,4,1,0]])
print(pred2)
```
